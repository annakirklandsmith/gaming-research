```python
import pandas
from pandas import Series, DataFrame
```

```python
dataA = pandas.read_excel('Teens_Gaming_2008.xlsx', 'Teens_Gaming_2008',)
dataA.head()
    
```

```python
dataE = dataA[['psraid','k9a','k9b','k9c', 'k9d', 'k19']].dropna()
dataE.head(10)
```

Throwing out some data because there are not more than 24 hours in a day

```python
dataE.sort_values(by=['k19'], ascending=False).head(10)
dataE = dataE[dataE['k19']<24]
dataE.sort_values(by=['k19'], ascending=False).head(10)
```

# Split data into sets for testing, training

```python
import numpy as np
dataE['subset'] = np.random.randint(1, 4, dataE.shape[0])
dataE.head(10)
```

```python
dataE.loc[dataE.k9a == 1, 'console'] = 1
dataE.loc[dataE.k9a != 1, 'console'] = 0
dataE.loc[dataE.k9b == 1, 'handheld'] = 1
dataE.loc[dataE.k9b != 1, 'handheld'] = 0
dataE.loc[dataE.k9c == 1, 'computer'] = 1
dataE.loc[dataE.k9c != 1, 'computer'] = 0
dataE.loc[dataE.k9d == 1, 'mobile'] = 1
dataE.loc[dataE.k9d != 1, 'mobile'] = 0
dataE = dataE.drop(['k9a', 'k9b', 'k9c', 'k9d'], axis=1)
dataE.head(5)
```

```python
e1 = dataE.loc[dataE['subset']==1]
e2 = dataE.loc[dataE['subset']==2]
e3 = dataE.loc[dataE['subset']==3]
e4 = dataE.loc[dataE['subset']==4]
e0 = e1
e0.head(5)
```
